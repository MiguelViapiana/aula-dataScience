```python
import matplotlib.pyplot as plt
import numpy as np

anos = [
    1872, 1890, 1900, 1920, 1940,
    1950, 1960, 1970, 1980, 1991,
    2000, 2010, 2022
]

populacoes = [
    9930478, 14333915, 17438434, 30635605, 41236315,
    51944397, 70992343, 94508583, 121150573, 146917459,
    169872856, 190755799, 203080756
]

# Cria o gráfico
plt.plot(anos, populacoes, label='População', color='grey', linewidth=2.5)

# Adiciona título e rótulos
plt.title('Crescimento populacional brasileiro')
plt.xlabel('Anos')
plt.ylabel('População')

# Adiciona uma legenda
plt.legend()

# Mostra o gráfico
plt.grid(axis = 'y')
plt.show()
savefig('graf1.png')


plt.barh(anos, populacoes, label='População', color='gray', height=7)
plt.title('Crescimento populacional brasileiro')
plt.xlabel('População')
plt.ylabel('Anos')
plt.legend()
plt.grid(axis= 'x')
plt.show()
```
